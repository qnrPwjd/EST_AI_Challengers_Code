### 3) 배열의 주요 속성

배열을 받으면 가장 먼저 **모양(shape)** 부터 확인하는 습관을 들입시다.


<details>
<summary> ---------- 💡 예시 코드 보기 ---------- </summary>

```python
prices_2d = np.array([
    [135.58, 137.71, 140.15],
    [189.51, 191.20, 193.42],
    [248.50, 251.30, 245.10]
])

print(prices_2d.shape)   # (3, 3)  — 3행 3열
print(prices_2d.ndim)    # 2       — 2차원
print(prices_2d.dtype)   # float64 — 저장된 값의 자료형
print(prices_2d.size)    # 9       — 전체 요소 개수
```

</details>

````python
# 코드 입력
prices_2d = np.array([
    [135.58, 137.71, 140.15],
    [189.51, 191.20, 193.42],
    [248.50, 251.30, 245.10]
])

print(prices_2d.shape)
print(prices_2d.ndim) 
print(prices_2d.dtype)
print(prices_2d.size)
````

### 5) 배열 연산 — 값 전체에 한 번에 적용 (브로드캐스팅)

NumPy의 진짜 매력이 여기서 시작됩니다. **배열 전체에 연산자를 그대로 적용**할 수 있습니다.
반복문이 사라지는 순간을 여러 상황으로 체험해봅시다.

> 브로드캐스팅(Broadcasting)은 크기가 다른 배열도 NumPy가 자동으로 크기를 맞춰 계산해 주는 기능입니다.


<details>
<summary> ---------- 💡 예시 코드 보기 ---------- </summary>

```python
# 예시 1: 주가 전체에 한 번에 계산

print(prices + 10)              # 모든 값에 10 더하기
print(prices * 1.1)             # 모든 값에 10% 가산 (수수료 계산 등)
print(prices - prices.mean())   # 평균과의 차이 (편차)
```

### 7) axis 정복 — 2차원 배열의 통계는 "방향"이 절반입니다

2차원 배열에서 `mean()`을 그냥 부르면 **전체 평균** 하나만 나옵니다.
"종목별 평균" 또는 "날짜별 평균"을 원한다면 **axis(축)** 를 지정해야 합니다. 초급자가 가장 많이 헷갈리는 지점이니 확실히 잡고 갑시다.

```
axis=0 : ↓ 세로 방향으로 계산 (열별 결과)  → "각 날짜별" 통계
axis=1 : → 가로 방향으로 계산 (행별 결과)  → "각 종목별" 통계
```

암기 팁: **axis는 "없어지는 방향"** 입니다. axis=0이면 행(세로)이 눌려 없어지고 열별 값만 남습니다.

```python
print(prices_2d.mean(axis=1))   # 종목별 평균 (행 방향 →) : 3개
print(prices_2d.mean(axis=0))   # 날짜별 평균 (열 방향 ↓) : 3개
```

```python
list_a = [1, 2, 3]
```

| 구분 | 파이썬 리스트 | NumPy 배열 |
|---|---|---|
| `*` 연산 | 반복(복제) | 각 요소에 곱셈 |
| `+` 연산 | 이어붙이기(연결) | 같은 위치끼리 덧셈 |
| 전체 값 계산 | 반복문/컴프리헨션 필요 | 연산자를 그대로 적용 가능 |

| 속도 (대량 데이터) | 상대적으로 느림 | 훨씬 빠름 |
| 자료형 제약 | 서로 다른 자료형 혼합 가능 | 동일한 자료형만 저장 (그래서 빠름) |

> 💡 **핵심 요약**: 리스트의 `+`, `*`는 "이어붙이기", "반복"이지만, NumPy 배열의 `+`, `*`는 수학적인 "덧셈", "곱셈"입니다. 숫자 데이터를 다룬다면 리스트보다 NumPy 배열이 훨씬 직관적이고 빠릅니다.


## ⚠️ 자주 하는 실수

> ⚠️ 리스트와 NumPy 배열의 `+`/`*` 연산 결과가 다르다는 것을 잊고 혼동하는 경우가 매우 흔합니다.

> ⚠️ 서로 다른 `shape`(모양)의 배열끼리 연산하려다 `ValueError: operands could not be broadcast together`가 나는 경우가 있습니다.

> ⚠️ `arr.shape`는 함수 호출이 아니라 속성이므로 괄호 없이 사용해야 합니다 (`arr.shape()` ❌, `arr.shape` ✅).

> ⚠️ axis=0과 axis=1을 반대로 지정해서 "종목별 평균"을 구하려다 "날짜별 평균"을 구하는 실수 — 결과의 **개수**를 보고 검산하세요 (종목이 3개인데 결과가 5개면 방향이 틀린 것).

> ⚠️ `np.where(조건)`처럼 값 2개를 빼먹으면 인덱스 튜플이 반환됩니다. 값을 바꾸려면 반드시 3개 인자를 채우세요.


<details>
<summary> ---------- 💡 예시 코드 보기 ---------- </summary>

```python

b = np.array([1, 2])
# print(a + b)   # ❌ ValueError: 모양이 다름 (3개 vs 2개)
print(a.shape, b.shape)   # 연산 전에 shape부터 확인하는 습관!
```

## 📌 핵심 정리

- `np.array()`로 배열 생성, `shape`/`ndim`/`dtype`/`size`로 속성 확인 — 배열은 **동일 자료형**만 저장
- 인덱싱/슬라이싱: 1차원은 `arr[i]`, 2차원은 `arr[행, 열]` (`:`는 "전부")
- 배열의 `+`, `*`는 리스트와 달리 실제 수학 연산 (같은 위치끼리 계산) — 환율, 할인, 수익률이 한 줄로
- 통계: `mean()`/`sum()`/`max()`/`min()`/`std()` + 위치를 찾는 `argmax()`/`argmin()`
- **axis=0은 ↓열별, axis=1은 →행별** — 결과 개수로 검산하는 습관
- 불리언 인덱싱 `arr[arr > 값]` = 조건으로 골라내기 — 다음 시간 Pandas 필터링의 원리
- `np.where(조건, 참값, 거짓값)` = 조건에 따라 값 바꾸기
- `np.sort()` 정렬, `reshape()` 모양 변경 (요소 개수 유지)

### 10) 서로 다른 자료형 비교 — Series vs DataFrame


<details>
<summary> ---------- 💡 예시 코드 보기 ---------- </summary>

```python
s = pd.Series([135.58, 189.51, 248.50])   # 1차원
print(type(s))       # <class 'pandas.core.series.Series'>

df2 = pd.DataFrame({"현재가": [135.58, 189.51, 248.50]})   # 2차원 (컬럼이 1개뿐이어도 DataFrame)
print(type(df2))     # <class 'pandas.core.frame.DataFrame'>

# DataFrame에서 컬럼 하나를 뽑으면 다시 Series가 됨
print(type(df2["현재가"]))   # <class 'pandas.core.series.Series'>
```

</details>

````python
# 코드 입력



````

| 구분 | Series | DataFrame |
|---|---|---|
| 차원 | 1차원 (하나의 열) | 2차원 (여러 열이 모인 표) |
| 비유 | 엑셀의 한 개 열 | 엑셀 시트 전체 |
| 생성 방법 | `pd.Series([...])` | `pd.DataFrame({...})` |
| DataFrame과의 관계 | DataFrame의 각 컬럼이 Series | 여러 Series가 모여서 구성됨 |

> 💡 **핵심 요약**: DataFrame에서 컬럼을 하나만 뽑으면 Series가 되고, 여러 Series를 옆으로 나란히 붙이면 DataFrame이 됩니다. 그리고 NumPy에서 배운 연산·통계·조건 선택의 원리가 Series에서 "이름표가 붙은 채로" 그대로 동작합니다.

**오늘의 3층 구조로 정리하면:**

```
NumPy 배열   : 값                     (위치로만 접근)
Series       : 값 + 인덱스             (이름으로 접근 가능)
DataFrame    : 여러 Series + 컬럼명    (표 전체)
```


## ⚠️ 자주 하는 실수

## 📌 핵심 정리

- Series는 1차원(인덱스+값), DataFrame은 2차원(여러 Series의 표) — Pandas = NumPy + 이름표
- Series 연산은 NumPy와 동일하되, Series끼리는 **같은 인덱스끼리** 계산
- DataFrame 생성: 딕셔너리(`{컬럼명: 값목록}`) / 리스트+`columns` / 딕셔너리의 리스트
- 컬럼 선택: `df["컬럼"]`(Series) vs `df[["컬럼"]]`(DataFrame)
- 컬럼 추가: `df["새컬럼"] = 값 또는 계산식` / 삭제: `df.drop("컬럼", axis=1)`
- 컬럼 통계: `mean()/sum()/max()` + 위치를 찾는 `idxmax()` (→ `df.loc[...]`로 행 확인)
- 정렬: `sort_values("컬럼", ascending=False)` — 원본은 그대로, 결과는 변수에 담기
- 조건 선택 맛보기: `df[df["컬럼"] >= 값]` — 본격 필터링은 다음 시간(CH07)에!
