### 6) 서로 다른 방식 비교 — `unique()` vs `value_counts()`

<details>
<summary> ---------- 💡 예시 코드 보기 ---------- </summary>

```python
dept = df["부서"]

print(dept.unique())         # ['개발팀' '마케팅팀' '인사팀']  ← "어떤 값들이 있는지" 목록만
print(dept.nunique())         # 3                              ← "몇 종류인지" 개수만
print(dept.value_counts())    # 개발팀 4 / 마케팅팀 3 / 인사팀 3  ← "각각 몇 번씩" 등장하는지
```

</details>

````python
# 코드 입력



````

| 함수 | 반환하는 것 | 반환 타입 |
|---|---|---|
| `unique()` | 중복 제거된 값 목록 | NumPy 배열 |
| `nunique()` | 고유값의 개수(숫자 하나) | 정수 |
| `value_counts()` | 값별 등장 횟수 | Series (값이 인덱스, 개수가 값) |

> 💡 **핵심 요약**: "어떤 값들이 있는지"만 궁금하면 `unique()`, "몇 종류인지 개수"만 필요하면 `nunique()`, "각 값이 얼마나 자주 나오는지"까지 알고 싶으면 `value_counts()`를 사용하세요.

### 10) 최댓값·최솟값의 "위치" 찾기 — `idxmax()`, `idxmin()`

<details>
<summary> ---------- 💡 예시 코드 보기 ---------- </summary>

```python
print(df["급여"].max())        # 최댓값 "값" 자체 → 6200
print(df["급여"].idxmax())     # 최댓값이 있는 "행 인덱스" → 5

best = df.loc[df["급여"].idxmax()]     # 급여가 가장 높은 직원의 전체 정보
print(best)

youngest = df.loc[df["나이"].idxmin()]  # 나이가 가장 어린 직원의 전체 정보
print(youngest)
```

</details>

````python
# 코드 입력



````

> 💡 `max()`/`min()`은 값 자체를, `idxmax()`/`idxmin()`은 그 값이 있는 **행의 인덱스**를 돌려줍니다. "급여가 가장 높은 사람이 **누구**인지"처럼 값이 아니라 해당 행 전체가 궁금할 때는 `df.loc[df["급여"].idxmax()]` 패턴으로 조합해서 사용합니다 (이번 소단원의 `loc[]` 없이도, 다음 소단원에서 배울 `loc[]`와 연결되는 대표 패턴입니다).

## ⚠️ 자주 하는 실수

> ⚠️ `describe()`를 기본값으로만 써서 문자열 컬럼의 통계(최빈값 등)를 놓치는 경우가 흔합니다 — `include='all'` 옵션을 기억하세요.

> ⚠️ `isnull()`과 `isna()`가 다른 함수라고 착각하는 경우가 있는데, 둘은 완전히 동일한 함수의 다른 이름(별칭)입니다.

> ⚠️ `df.isnull()`만 보고 각 셀의 True/False를 눈으로 세려고 하는 경우가 있는데, `.sum()`을 붙여 컬럼별 개수로 요약해서 보는 것이 훨씬 효율적입니다.

> ⚠️ `sort_values()` 같은 함수는 기본적으로 원본을 바꾸지 않고 "새 결과"를 반환합니다. 계속 사용하려면 `df = df.sort_values(...)`처럼 재할당하거나 `inplace=True`를 사용해야 합니다.

> ⚠️ `idxmax()`가 돌려주는 것은 "값"이 아니라 "인덱스"입니다. 최댓값 자체가 필요하면 `max()`를 사용해야 합니다.

## 🚀 실무에서는?

> 🚀 새로운 데이터를 받으면 실무자는 거의 항상 `df.head()` → `df.info()` → `df.describe(include='all')` → `df.isnull().sum()` 순서로 훑어보며 데이터의 전반적인 상태를 파악합니다. 이후 `sort_values()`/`nlargest()`로 상위·하위 값을 확인하고, `idxmax()`로 "1등이 누구인지"까지 짚어보면 EDA의 "국룰" 루틴이 하나로 완성됩니다.

| 구분 | `loc[]` | `iloc[]` |
|---|---|---|
| 기준 | 인덱스 라벨(이름) | 정수 위치(순서) |
| 슬라이싱 끝 포함 여부 | **포함** (`0:2`는 0,1,2) | **미포함** (`0:2`는 0,1만) |

| 인덱스가 문자열일 때 | 그 문자열로 바로 접근 가능 | 여전히 숫자 위치로만 접근 |

> 💡 **핵심 요약**: 이름표(라벨)로 찾고 싶으면 `loc`, 순서(몇 번째)로 찾고 싶으면 `iloc`을 사용하세요. 특히 **슬라이싱 시 끝 포함 여부가 다르다**는 점이 가장 헷갈리는 포인트이니 꼭 기억하세요.

```python
print(df.loc[0])
```

>
> ⚠️ 컬럼명에 공백이나 특수문자가 있으면 백틱(`` ` ``)으로 감싸야 합니다. 예: `` df.query("`부서 명` == '개발팀'") ``

## ⚠️ 자주 하는 실수

> ⚠️ `loc`와 `iloc`의 슬라이싱 끝 포함 여부를 혼동해서 예상보다 한 개 많거나 적은 행을 가져오는 경우가 매우 흔합니다.

> ⚠️ 다중 조건에서 `and`/`or`를 그대로 써서 `ValueError: The truth value of a Series is ambiguous`가 발생하는 경우가 흔합니다. (&, |)

> ⚠️ 조건을 괄호로 감싸지 않아서 연산자 우선순위 문제로 예상과 다른 결과가 나오는 경우가 있습니다.

> ⚠️ `query()`에서 외부 변수를 `@` 없이 그대로 쓰면 컬럼명으로 착각해 에러가 나거나, 문자열 값에 따옴표를 빠뜨려 문법 오류가 나는 경우가 흔합니다.

<details>
<summary> ---------- 💡 예시 코드 보기 ---------- </summary>

```python
df[(df["부서"] == "개발팀") & (df["급여"] >= 4000)]    # ✅ 각 조건을 괄호로 감쌈
```

## 📌 핵심 정리

- 컬럼 선택: `df["컬럼"]`(Series) vs `df[["컬럼"]]`(DataFrame)
- `loc[]`: 라벨 기준, 슬라이싱 끝 포함 / `iloc[]`: 위치 기준, 슬라이싱 끝 미포함
- 조건 필터링: `df[조건]`, 다중 조건은 `&`(and)/`|`(or)/`~`(not) + 괄호 필수
- `isin()`: 여러 값 중 일치 확인 / `between()`: 범위 확인 / `str.contains()`: 문자열 포함 확인
- 단일 값 접근: `at[]`(라벨), `iat[]`(위치) — 값 하나를 빠르게 읽거나 수정
- 인덱스 관리: `set_index()`(의미 있는 컬럼을 인덱스로), `reset_index()`(기본 인덱스로 복원)
- `query()`: 조건을 문자열로 작성, `and`/`or` 사용 가능, 외부 변수는 `@`로 참조

### 6) 서로 다른 방식 비교 — `groupby()+agg()` vs `pivot_table()`

<details>
<summary> ---------- 💡 예시 코드 보기 ---------- </summary>

```python
# 방법 A: groupby() + agg()

print(result_a)

# 방법 B: pivot_table()
result_b = df.pivot_table(index="부서", values="급여", aggfunc=["mean", "count"])
print(result_b)
```

</details>

````python
# 방법 A: groupby() + agg()




# 방법 B: pivot_table()



````

| 구분 | `groupby()` + `agg()` | `pivot_table()` |
|---|---|---|
| 결과 형태 | 그룹 기준이 인덱스인 표 | 엑셀 피벗테이블과 유사한 표 (행/열 기준을 각각 지정 가능) |
| 행과 열 모두 기준 지정 | 불편함 (컬럼 하나만 인덱스로) | `index`와 `columns`를 각각 지정해서 2차원 표로 정리 가능 |
| 사용 난이도 | 비교적 간단하고 직관적 | 옵션이 많아 유연하지만 살짝 더 복잡 |
| 언제 사용? | 하나의 기준으로만 그룹화할 때 | 두 개 이상의 범주를 행/열로 교차해서 보고 싶을 때 (엑셀 피벗과 비슷한 리포트) |

> 💡 **핵심 요약**: 단순히 "하나의 기준으로 그룹화해서 통계 내기"는 `groupby()`가 간편하고, "두 범주를 행/열로 교차시킨 표"가 필요하면 `pivot_table()`이 더 적합합니다. 실제로 `pivot_table()`은 내부적으로 `groupby()`를 활용해서 만들어져 있습니다.

### 8) 구간으로 나눠서 그룹화 — `cut()`, `qcut()`

```python
df["나이대"] = pd.cut(df["나이"], bins=[0, 30, 40, 100],
                     labels=["30세 이하", "31~40세", "41세 이상"])
print(df.groupby("나이대", observed=True)["급여"].mean())

df["급여등급"] = pd.qcut(df["급여"], q=3, labels=["하", "중", "상"])
```

### 9) 교차표 심화 — `crosstab()`을 비율로 보기 (`normalize`)

<details>
<summary> ---------- 💡 예시 코드 보기 ---------- </summary>

```python
print(pd.crosstab(df["부서"], df["도시"]))                        # 기본: 빈도수(개수)
print(pd.crosstab(df["부서"], df["도시"], normalize="index"))     # 행(부서) 기준 비율
print(pd.crosstab(df["부서"], df["도시"], normalize="columns"))   # 열(도시) 기준 비율
```

</details>

````python
# 코드 입력



````

> 💡 `normalize`는 개수 대신 비율로 바꿔줍니다. `normalize="index"`는 "각 부서 안에서 도시별 비중", `normalize="columns"`는 "각 도시 안에서 부서별 비중"으로, 기준 방향이 달라지면 해석도 완전히 달라지니 어느 쪽을 100%로 보는지 항상 확인하세요.
>
> 합계 행/열을 추가하는 `margins=True` 옵션도 있는데, 이는 CH09(데이터 병합)의 `pivot_table()` 심화에서 자세히 다룹니다.

> ⚠️ `crosstab(normalize=...)`에서 `"index"`와 `"columns"`의 기준 방향을 헷갈려 비율 해석을 반대로 하는 경우가 있습니다.

## 📌 핵심 정리

- `groupby(기준)`으로 그룹을 나눈 뒤 `.mean()`, `.sum()` 등 집계 함수를 붙여야 계산됨
- `agg(["함수1","함수2"])` 또는 `agg({"컬럼":"함수"})`로 여러 통계를 한 번에
- `agg(새컬럼명=("원본컬럼","함수"))` Named Aggregation으로 결과 컬럼명을 바로 지정
- `transform()`: 그룹 계산 결과를 원본 행 개수 그대로 유지 (새 컬럼 추가에 유용)
- `filter()`: 조건을 만족하는 그룹 전체를 선택
- `cut()`/`qcut()`: 연속값을 구간(범주)으로 변환 → `groupby()`와 연결해 구간별 비교
- `crosstab()`: 빈도표, `normalize="index"/"columns"`로 비율 확인 / `pivot_table()`: 통계 계산이 가능한 유연한 교차표
