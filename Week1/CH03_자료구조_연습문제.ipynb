### 문제 16. (중급) keys() / values() / items() 순회
`menu` 딕셔너리의 키, 값, 키-값 쌍을 각각 리스트로 출력하고, `items()`로 순회하며 출력하세요.

```python
menu = {"아메리카노": 4500, "카페라떼": 5000, "에이드": 5500}

# list(menu.keys())   로 키 목록을 출력하시오.
print(list(menu.keys()))
# list(menu.values()) 로 값 목록을 출력하시오.
print(list(menu.values()))
# list(menu.items())  로 키-값 쌍 목록을 출력하시오.
print(list(menu.items()))

# for item, price in menu.items(): 로 순회하며
# f"{item}: {price:,}원" 형태로 출력하시오.
for item, price in menu.items():
    print(f"{item}: {price}원")
```

### 문제 18. (중급) 딕셔너리 정렬
`scores` 딕셔너리를 값(점수) 기준으로 내림차순 정렬하여 순위를 출력하세요.

```python
scores = {"김민준": 88, "이서연": 95, "박지훈": 72, "최유나": 91}

# 힌트: sorted(scores.items(), key=lambda x: x[1], reverse=True)
# 값(점수) 기준 내림차순으로 sorted_desc에 저장하시오.
sorted_desc = sorted(scores.items(), key = lambda x: x[1], reverse = True)

# for rank, (name, score) in enumerate(sorted_desc, start=1): 로 순위를 출력하시오.
# 출력 형식: "1위: 이서연 (95점)"
for rank, (name, score) in enumerate(sorted_desc, start = 1):
    print(f"{rank}위: {name}({score}점)")
```

### 문제 19. (중급) 딕셔너리 컴프리헨션
`names`와 `scores` 리스트를 `zip`으로 묶어 딕셔너리를 만들고, 80점 이상인 사람만 필터링하세요.

```python
names = ["김민준", "이서연", "박지훈"]
scores = [88, 95, 72]

# 힌트: {name: score for name, score in zip(names, scores)}
# zip으로 묶어 딕셔너리 컴프리헨션으로 score_dict를 만들고 출력하시오.
print({name: score for name, score in zip(names, scores)})
# 조건부 컴프리헨션으로 80점 이상인 사람만 high_scorers에 저장하고 출력하시오.
high_scorers = {name: score for name, score in zip(names, scores) if score >= 80}
print(high_scorers)
```

```python
student_a = {"파이썬", "자료구조", "영어회화"}
student_b = {"자료구조", "운영체제", "영어회화", "선형대수"}

# 힌트: 합집합 | / 교집합 & / 차집합 - / 대칭차집합 ^
# student_a | student_b 로 합집합을 출력하시오.
print(student_a | student_b)
# student_a & student_b 로 교집합을 출력하시오.
print(student_a & student_b)
# student_a - student_b 로 차집합(A에만 있는 것)을 출력하시오.
print(student_a - student_b)
# student_a ^ student_b 로 대칭차집합(한쪽에만 있는 것)을 출력하시오.
print(student_a ^ student_b)
```

### 문제 24. (중급) issubset / issuperset / isdisjoint
