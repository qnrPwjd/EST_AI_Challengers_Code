### 4) 주요 리스트 메서드

| 메서드 | 역할 | 예시 |
|---|---|---|
| `.append(x)` | 맨 뒤에 요소 1개 추가 | `[1,2].append(3)` → `[1,2,3]` |
| `.insert(i, x)` | i번 위치에 요소 삽입 | `[1,3].insert(1,2)` → `[1,2,3]` |
| `.remove(x)` | 값이 x인 첫 요소 삭제 | `[1,2,3].remove(2)` → `[1,3]` |
| `.pop(i)` | i번 요소를 꺼내면서 삭제 (기본은 마지막) | `[1,2,3].pop()` → `3` 반환, 리스트는 `[1,2]` |
| `.sort()` | 오름차순 정렬 (원본 변경) | `[3,1,2].sort()` → `[1,2,3]` |
| `.reverse()` | 순서 뒤집기 (원본 변경) | `[1,2,3].reverse()` → `[3,2,1]` |
| `.index(x)` | 값 x의 첫 위치 반환 | `[1,2,3].index(2)` → `1` |
| `.count(x)` | 값 x의 개수 반환 | `[1,2,2,3].count(2)` → `2` |

<details>
<summary> ---------- 💡 예시 코드 보기 ---------- </summary>

```python
nums = [3, 1, 4, 1, 5]

nums.append(9)
print(nums)          # [3, 1, 4, 1, 5, 9]

nums.insert(0, 100)
print(nums)          # [100, 3, 1, 4, 1, 5, 9]

nums.remove(1)        # 값이 1인 "첫 번째" 요소만 삭제
print(nums)          # [100, 3, 4, 1, 5, 9]

popped = nums.pop()
print(popped, nums)   # 9  [100, 3, 4, 1, 5]

nums.sort()
print(nums)          # [1, 3, 4, 5, 100]

nums.reverse()
print(nums)          # [100, 5, 4, 3, 1]
```

### 5) 서로 다른 방식 비교 — `append()` vs `extend()`

둘 다 "리스트에 값을 추가"하는 메서드지만 동작이 완전히 다릅니다.

<details>
<summary> ---------- 💡 예시 코드 보기 ---------- </summary>

```python
a = [1, 2, 3]
b = [1, 2, 3]

a.append([4, 5])     # 리스트 자체를 "하나의 요소"로 통째로 추가
print(a)               # [1, 2, 3, [4, 5]]

b.extend([4, 5])      # 리스트 안의 "각 요소"를 하나씩 풀어서 추가
print(b)               # [1, 2, 3, 4, 5]
```

</details>

````python
# 코드 입력


````

| 메서드 | 동작 | 결과 예시 |
|---|---|---|
| `append(x)` | x를 통째로 하나의 요소로 추가 | `[1,2,3].append([4,5])` → `[1,2,3,[4,5]]` |
| `extend(x)` | x 안의 요소들을 낱개로 풀어서 추가 | `[1,2,3].extend([4,5])` → `[1,2,3,4,5]` |

> 💡 **핵심 요약**: 리스트를 "하나의 덩어리"로 넣고 싶으면 `append()`, 리스트 안의 값들을 "낱개로 풀어서" 이어붙이고 싶으면 `extend()`를 사용합니다.

### 3) 서로 다른 자료형 비교 — List vs Tuple

겉보기엔 비슷하지만, 언제 어떤 것을 쓸지는 명확한 기준이 있습니다.

| 구분 | List | Tuple |
|---|---|---|
| 생성 기호 | `[]` | `()` |
| 변경 가능 여부 | 가능 (mutable) | 불가능 (immutable) |

### 3) 주요 딕셔너리 메서드

| 메서드 | 역할 | 예시 |
|---|---|---|
| `.keys()` | 모든 키 목록 | `{"a":1}.keys()` → `dict_keys(['a'])` |
| `.values()` | 모든 값 목록 | `{"a":1}.values()` → `dict_values([1])` |
| `.items()` | (키,값) 쌍 목록 | `{"a":1}.items()` → `dict_items([('a',1)])` |
| `.get(key)` | 값 조회 (없으면 None) | `{"a":1}.get("b")` → `None` |
| `.pop(key)` | 키를 삭제하며 값 반환 | `{"a":1}.pop("a")` → `1` |
| `.update(dict2)` | 다른 딕셔너리 내용으로 갱신 | 아래 예시 참고 |
| `.setdefault(key, 기본값)` | 키가 없으면 기본값으로 추가 | 아래 예시 참고 |

<details>
<summary> ---------- 💡 예시 코드 보기 ---------- </summary>

```python
person = {"name": "홍길동", "age": 25}

print(person.keys())     # dict_keys(['name', 'age'])
print(person.values())   # dict_values(['홍길동', 25])
print(person.items())    # dict_items([('name', '홍길동'), ('age', 25)])

age = person.pop("age")
print(age, person)        # 25 {'name': '홍길동'}

person.update({"age": 30, "city": "부산"})
print(person)              # {'name': '홍길동', 'age': 30, 'city': '부산'}

person.setdefault("job", "무직")   # "job" 키가 없으므로 추가됨
print(person)
person.setdefault("name", "이몽룡")  # "name" 키가 이미 있으므로 변경 안 됨
print(person)
```

### 4) 서로 다른 조회 방식 비교 — `dict[key]` vs `.get(key)`

같은 "값을 조회"하는 목적이지만, 키가 없을 때의 동작이 완전히 다릅니다.

<details>
<summary> ---------- 💡 예시 코드 보기 ---------- </summary>

```python
person = {"name": "홍길동", "age": 25}

print(person["name"])        # 홍길동 (정상 조회)
# print(person["job"])        # ❌ KeyError! "job" 키가 없어서 에러 발생

print(person.get("job"))      # None (에러 없이 None 반환)
print(person.get("job", "무직"))   # "무직" (기본값을 직접 지정 가능)
```

</details>

````python
# 코드 입력


````

| 방식 | 키가 없을 때 | 언제 사용? |
|---|---|---|
| `dict[key]` | `KeyError` 발생 (프로그램 중단 위험) | 키가 반드시 존재한다고 확신할 때 |
| `dict.get(key)` | `None` 반환 (안전) | 키가 없을 수도 있는 경우 |
| `dict.get(key, 기본값)` | 지정한 기본값 반환 | 없을 때 대체값이 필요한 경우 |

> 💡 **핵심 요약**: 키가 확실히 있다는 보장이 없다면 `[]`보다 `.get()`을 사용하는 것이 더 안전합니다.

# 📘 CH03-04. Set

## 🤔 먼저 생각해보기

> **설문조사 응답에서 "중복 없이 어떤 답변들이 나왔는지"만 알고 싶다면 어떻게 해야 할까요?**

리스트에 값을 그대로 모으면 같은 답변이 여러 번 중복돼서 쌓입니다. "중복은 자동으로 제거하고, 어떤 값들이 있었는지"만 알고 싶을 때 사용하는 자료구조가 **셋(Set)**입니다.

---

## 📖 핵심 개념

### 1) 셋이란? — 중복 없는 순서 없는 값의 모음

- 중괄호 `{}`로 생성 (딕셔너리와 기호는 같지만 `키:값` 없이 값만 나열)
- **중복을 자동으로 제거**
- **순서가 없어서** 인덱싱/슬라이싱 불가능

<details>
<summary> ---------- 💡 예시 코드 보기 ---------- </summary>

```python
fruits = {"사과", "바나나", "사과", "포도"}

numbers = set([1, 2, 2, 3, 3, 3])
print(numbers)          # {1, 2, 3}
```

### 3) 서로 다른 삭제 방식 비교 — `remove()` vs `discard()`

둘 다 값을 삭제하지만, **없는 값을 삭제하려 할 때** 동작이 다릅니다.

<details>
<summary> ---------- 💡 예시 코드 보기 ---------- </summary>

```python
fruits = {"사과", "바나나", "포도"}

fruits.remove("바나나")     # 정상적으로 삭제됨
print(fruits)

# fruits.remove("수박")     # ❌ KeyError! 없는 값을 remove하면 에러 발생

fruits.discard("수박")       # 에러 없이 그냥 무시됨 (수박이 없어도 안전)
print(fruits)
```

</details>

````python
# remove 활용

````

````python
# discard 활용


````

| 메서드 | 값이 없을 때 | 언제 사용? |
|---|---|---|
| `.remove(x)` | `KeyError` 발생 | 값이 반드시 있다고 확신할 때 |
| `.discard(x)` | 에러 없이 무시 | 값이 없을 수도 있는 경우 (더 안전) |

> 💡 **핵심 요약**: 딕셔너리의 `[]` vs `.get()` 비교와 똑같은 원리입니다 — 확실하지 않으면 에러 없는 쪽(`discard`)을 사용하세요.

### 2) 서로 다른 방식 비교 — 수동 인덱스 관리 vs `enumerate()`

`enumerate()` 없이도 인덱스를 얻을 수는 있지만, 훨씬 번거롭고 실수하기 쉽습니다.

<details>

### 5) 서로 다른 방식 비교 — for문 vs List Comprehension

같은 결과를 만드는 두 가지 방법을 비교해봅시다.

<details>
<summary> ---------- 💡 예시 코드 보기 ---------- </summary>

```python
# [버전 A] 일반 for문 (4줄)
result = []

    if i % 2 == 0:
        result.append(i ** 2)
print(result)   # [4, 16, 36, 64, 100]
```

</details>

````python
# 코드 입력
# [버전 A] 일반 for문 (4줄)


````

<details>
<summary> ---------- 💡 예시 코드 보기 ---------- </summary>

```python
# [버전 B] 리스트 컴프리헨션 (1줄)
result = [i ** 2 for i in range(1, 11) if i % 2 == 0]
print(result)   # [4, 16, 36, 64, 100]
```

</details>

````python
# 코드 입력
# [버전 B] 리스트 컴프리헨션 (1줄)

````

| 구분 | 코드 길이 | 가독성 | 언제 사용? |
|---|---|---|---|
| for문 | 김 (여러 줄) | 로직이 복잡할 때 오히려 더 읽기 쉬움 | 조건/처리가 복잡하거나 여러 단계일 때 |
| List Comprehension | 짧음 (한 줄) | 간단한 변환/필터링에 직관적 | 하나의 리스트를 간단히 만들 때 |

> 💡 **핵심 요약**: 로직이 단순하면 컴프리헨션이 간결하고 좋지만, 조건이 여러 겹으로 복잡해지면 오히려 일반 for문이 더 읽기 쉽습니다. 무조건 짧게 쓰는 것이 능사는 아닙니다.

print(total)

print(total2)

## 📌 핵심 정리

- 기본 문법: `[표현식 for 변수 in 반복대상]`
- 필터링: `[표현식 for 변수 in 반복대상 if 조건]` (if는 뒤)
- 값 변환: `[값1 if 조건 else 값2 for 변수 in 반복대상]` (if-else는 앞)
- for문 대비 간결하지만, 로직이 복잡해지면 오히려 일반 for문이 더 읽기 쉬움
- 중첩 for도 한 줄로 표현 가능 (평탄화 등에 활용)
